## Declaring the size of an array
Arrays  are different to the simpler data types we have mainly seen before in that their size and so the amount of space they need varies from array to array. As with records the space needs to be allocated for them before they are used. If we initialise them at the same time as declaring them then this can be sorted out automatically by the compiler. If not then it is a good idea to allocate the space at declaration time. 

We use keyword **new** to do this following it with the Array type but with the size  of the array in the square brackets.

A typical declaration is therefore:

```java
String [] a = new String[100];
```
This declares a String array a of size 100 - it allocates space in a for an array holding 100 String values.

## Summary
The storage space for an array must be explicitly allocated, so a size must be given when the array is declared as in:
```java
String [] a = new String[1000];
```
which allocates 1000 storage spaces in array a. Once set the size of an array cannot be changed.

We can process arrays using counter-controlled for loops. The for loop just steps throught the array, doing what needs to be done on each element of the array. This might be setting values, printing values, or any other action that needs to be done on every element of the array.

Such a for loop can adapt to the size of the array if the size is stored in a variable. We can also split off operations to process an array into a separate method. SUch methods can be very general, such as a create array or set array method.
